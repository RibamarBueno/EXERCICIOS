```python

def escolha_servico():

  print('----------BEM VINDO, A CÓPIAS RIBAMAR BUENO------------')
  print("................. MENU SERVIÇO ........................")
  while True:
    print("--|CÓDIGO|---------SERVIÇO--------|----VALORES----|--")
    print("--| DIG  |DIGITAÇÃO               |    R$:1,10    |--")
    print("--| ICO  |IMPRESSÃO COLORIDA      |    R$:1,00    |--")
    print("--| IBO  |IMPRESSÃO PRETO E BRANCO|    R$:0,40    |--")
    print("--| FOT  |FOTOCÓPIA               |    R$:0,20    |--")
    serv=input("Digite o código do serviço desejado:\n")
    serv=serv.upper()

    if serv != "DIG" and serv != "ICO" and serv != "IBO" and serv != "FOT":
      print("Digite um código válido.")
      continue

    if serv == "DIG":
      print("Este serviço tem o custo de 1,10, por cópia.")
      return 1.10
      break

    elif serv == "ICO":
      print("Este serviço tem o custo de 1,00, por cópia.")
      return 1.00
      break

    elif serv== "IBO":
      print("Este serviço tem o custo de 0,40, por cópia.")
      return 0.40
      break

    elif serv=="FOT":
      print("Este serviço tem o custo de 0,20, por cópia.")
      return 0.20
      break

def num_paginas():
  print(".................NÚMERO DE PAGINAS.....................")
  while True:
    np=int(input('Digite o número de páginas:\n'))
    if np > 10000:
          print("Não é aceito pedidos nessa quantidade de páginas.")
          continue

    elif 10 <= np <= 100:
          desc=0.10
          print("Você tem direito a 10% de desconto")
          return np
          break

    elif 100 <= np <= 1000:
          desc=0.15
          print("Você tem direito a 15% de desconto")
          return np
          break

    elif 1000 <= np < 10000:
          desc=0.20
          print("Você tem direito a 20% de desconto")
          return np
          break


def servico_extra():
  while True:

    print("-------------------ADICIONAIS------------------------")
    print("--|CÓDIGO|---------SERVIÇO--------|----VALORES----|--")
    print("--|1    |ENCADERNAÇÃO SIMPLES     |     R$: 10,00 |--")
    print("--|2    |ENCADERNAÇÃO DE CAPA DURA|     R$: 25,00 |--")
    print("--|3    |NÃO QUERO ADICIONAIS     |     R$:  0,00 |--")

    extra=input("Digite o serviço adicional:\n")
    if extra != "1" and extra != "2" and extra != "3" :
      print("Digite um código válido.")
      continue
    if extra == "3":
      return 0.00
    elif extra == "1":
      return 10.00
    elif extra == "2":
      return 25.00

servico = escolha_servico()
#print("Valor :",servico)

np = num_paginas()
#print("Páginas :",np)

extra = servico_extra()
#print("Extra :", extra)

while True:
    if 10 <= np <= 100:
          desc=0.10
          break

    elif 100 <= np <= 1000:
          desc=0.15
          break

    elif 1000 <= np < 10000:
          desc=0.20
          break

vt=np*servico
desct=vt*desc
valortotal=vt-desct
valorextra=valortotal+extra

print('Total R$:{}\nServiço: {}\nPáginas: {}\nExtra: {}'.format(valorextra,servico,np,extra))
```
