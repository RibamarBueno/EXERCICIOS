```python
#--------------------------inicio das variáveis globais------------------------#

lista_livro =[]
id_global = 0 #NOSSO CONTADOR

#--------------------------final das variáveis globais-------------------------#

#--------------------------INICIO CADASTRAR LIVRO ------------------------#
def cadastrar_livro(id):
  print('-------MENU DE CADASTRO-------')
  print('ID do livro:{}'.format(id))#ID CONTADOR
  nome=                     input('Digite o NOME do livro:\n'    )
  autor=                    input('Digite o AUTOR do livro.\n'   )
  editora=                  input('Digite a EDITORA do livro.\n')
  dicionario_livro = {'id' :id,
                      'nome'   :nome,
                      'autor'  :autor,
                      'editora'  :editora}
  lista_livro.append(dicionario_livro.copy())#"CRIAÇÃO DO DICIONÁRIO"


#--------------------------FINAL CADASTRAR LIVRO------------------------#

#--------------------------INICIO CONSULTAR LIVRO------------------------#
def consultar_livro():
  print('------------------MENU CONSULTAR LIVRO----------------------')
  while True:
    opcao_consultar = input('Escolha a opção desejada:\n'+
                            '1-Consultar TODOS os livros.\n'+
                            '2-Consultar livro por ID.\n'+
                            '3-Consultar livro por AUTOR.\n'+
                            '4-Retonar\n')
    if opcao_consultar == '1' :
      print('Você escolheu a opção consultar TODOS os livros.')
      for livro in lista_livro:
        print('-----------------------------------')
        for key,value in livro.items():
          print('{}: {}'.format(key,value))
        print('-----------------------------------')

    elif opcao_consultar == '2' :
      print('Você escolheu a opção consultar livro por ID.')
      valor_desejado = int(input('Digite o ID desejado:\n'))
      for livro in lista_livro:
        if livro ['id'] == valor_desejado:
          print('---------------------------------')
          for key,value in livro.items():
            print('{}: {}'.format(key,value))
          print('---------------------------------')

    elif opcao_consultar == '3' :
      print('Você escolheu a opção consultar livro por AUTOR.')
      valor_desejado = input('Digite com o AUTOR desejado:')
      for livro in lista_livro:
        if livro ['autor']  == valor_desejado:
          print('---------------------------------')
          for key,value in livro.items():
            print('{}: {}'. format(key,value))
        print('---------------------------------')

    elif opcao_consultar == '4' :
      return

    else:
      print('Opção inválida. Tente novamente.')
      continue
#-------------------------FINAL CONSULTAR LIVRO-------------------------#

#--------------------------INICIO REMOVER LIVRO------------------------#
def remover_livro():
  print('------------------MENU REMOVER LIVRO--------------------')
  valor_desejado = int(input('Digite o ID do livro que deseja remover:\n'))
  for livro in lista_livro:
    if livro['id'] == valor_desejado:
      lista_livro.remove(livro)
      print('Livro removido.')

#--------------------------FINAL REMOVER LIVRO------------------------#

#--------------------------INICIO DO MAIN-----------------------#
print('Bem-vindo, a livraria do Ribamar Bueno.')
while True:
  opcao_principal = input('Escolha a opção desejada:\n'+
                          '1- Cadastrar Livro.\n'+
                          '2- Consultar Livro.\n'+
                          '3- Remover Livro.\n'+
                          '4- Encerrar programa.\n')
  if opcao_principal =='1' :
    id_global = id_global + 1
    cadastrar_livro(id_global)
  elif opcao_principal == '2' :
    consultar_livro()
  elif opcao_principal == '3' :
    remover_livro()
  elif opcao_principal == '4' :
    break
  else:
    print('Opção inválida. Tente novamente.')
    continue

#--------------------------FINAL DO MAIN------------------------#
```
